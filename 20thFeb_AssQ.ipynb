```python
ANS-1-

both GET and POST mean sending data to server but the difference is that in the GET method,we send the data through the url like the input or some data that needs to be processed is sent via the url but in the POST method,data is sent via the body.

So,GET requests data from a speceficd resource and appends data to URL query string whereas the data that POST sends is included in the body of the request and not in the URL and the body can contain different types of data depending on the request,usually encoded in a JSON,XML format and when the server recieves this data,it extracts data as required.
```

```python
ANS-2-

Request is used in flask to acess the request data sent by the server so that we can access the data inside the request made and take the input and calculate the output and return it to the client
```
