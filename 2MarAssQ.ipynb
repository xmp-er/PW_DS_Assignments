```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#ANS-1-Matplot lib is a data visualization tool in python,it is used to visualize data and is helpful for statistical analysis,five modules that can be plotted using matplotlib are 
#Line plot
#Scatter plot
# Bar chart
# Histogram
# Pie chart
```

```python
#ANS-2-Scatter plot is used to generate data as points where the first argument elements are plotted in x-axis and the second argument elements are plotted in y-axis 
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("x-axis data")
plt.ylabel("y-axis data")
plt.title("Scatter plot example")
plt.plot(x,y)
```

```python
#ANS-3-Sublot function is used to plot multiple lines in the same figure for comparison
x1= np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2= np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3= np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4= np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plo=plt.figure()
a1=plo.add_subplot(2,2,1)
a1.plot(x1,y1)
a2=plo.add_subplot(2,2,2)
a2.plot(x2,y2)
a3=plo.add_subplot(2,2,3)
a3.plot(x3,y3)
a4=plo.add_subplot(2,2,4)
a4.plot(x4,y4)
plo.tight_layout()
```

```python
#ANS-4-Bar Plot is used to represent the data in a bar form where the first argument elements are in x-axis and second argument elements in y-axis and the data at same indexes are mapped to each other to repesent the data in a bar format
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.show()
plt.barh(company,profit)
plt.show()
```

```python
#ANS-5-Box plots are used to summarize and compare the distribution of a dataset across different groups or categories. They are particularly useful for identifying outliers, skewness, and the spread of the data. Box plots can also be used to detect differences in the median and IQR between different groups or categories.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
plt.boxplot(data)
```
